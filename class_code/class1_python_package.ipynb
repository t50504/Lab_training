```python
drop_value = data.dropna() # 有遺失值的觀測值都刪除  
print(drop_value)  
print("---")  
filled_value = data.fillna(0) # 有遺失值的觀測值填補 0  
print(filled_value)  
print("---")  
filled_value_column = data.fillna({"student": "NULL", "english": 0}) # 依欄位填補遺失值  
print(filled_value_column)  
```

```python
english_value = data[data['english'].notnull()]
science_value = data[data['english'].notnull()]
print(english_value)
```

```python
eng_avg = sum(list(english_value['english'])) / len(english_value)
sc_avg = sum(list(science_value['english'])) / len(science_value)
```

```python
data = data.fillna({"student": "NULL", "english": eng_avg ,"science": sc_avg})
print(data)
```

```python
# iteration
# method1 for loop
total = []
for i in range(len(data)):
    total.append(data['math'][i] + data['english'][i] + data['science'][i])
    
print(total)
data['total score1'] = total
```

```python
#method 2 iterrows
result=[]
for i,row in data.iterrows():
    result.append(row['math'] + row['english'] + row['science'])
data['total score2'] = result
```
